```python
#all values are zero, this indicates no null values are present in the dataset
#checking for duplicate data, if present then drop them
data_dup=data.duplicated().any()
print(data_dup)
```

```python
#we can see most people with disease had age 50-60
#checking chest pain type(Value 0,1,2 or 3)
sns.countplot(data['cp'])
plt.xticks([0,1,2,3],["typical angina","atypical angina","non-anginal pain","asymptomatic"])
plt.show()
```

```python
#fasting blood sugar distribution according to target variable
sns.countplot(x="fbs",hue="target",data=data)
plt.legend(labels=["No disease","Disease"])
plt.show()
```

```python
#plotting the continuous variables
#separating columns having continuous values and columns having categorical values
data.columns
```

```python
cate_val=[]#categorical
cont_val=[]#continuous

for column in data.columns:
    if data[column].nunique()<=10:#returns number of unique values in the column
        cate_val.append(column)
    else:
        cont_val.append(column)
```

```python
cate_val
```

```python
cont_val
```

```python
#plotting continuous value
data.hist(cont_val,figsize=(15,6))
plt.tight_layout()
plt.show()
```
